```python
# a. Using sharpening Mask
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = r'/home/hansel/ip/Images/image1.jpg'
img = cv2.imread(path)  
filter=np.array([[-1,-1,-1],
                     [-1,9,-1],
                     [-1,-1,-1]])
sharpen_img=cv2.filter2D(img,-1,filter)
plt.subplot(121),plt.title("Original"),plt.imshow(img)
plt.subplot(122),plt.title("sharpen image"),plt.imshow(sharpen_img)   
```

```python
# b. 2-D laplacian high pass filter in spatial domain
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = r'/home/hansel/ip/Images/image1.jpg'
img = cv2.imread(path,0) 
kernel1=np.array([[0,0,0],
                  [0,1,0],
                  [0,0,0]])
blur=cv2.filter2D(img,-1,kernel1)
laplacian=cv2.Laplacian(blur,5,cv2.CV_64F)
filtered_image=cv2.convertScaleAbs(laplacian)
plt.figure(figsize=(4,5))
plt.imshow(filtered_image,cmap="gray")
plt.title('laplacian')
plt.show()
```
