```python
import cv2

image1 = cv2.imread(r"E:\Images\cat2.jpg")
image2 = cv2.imread(r"E:\Images\car2.jpg")
img1 = cv2.resize(image1,(400,500))
img2 = cv2.resize(image2,(400,500))

cv2.imshow("Orginal_img1",img1)
cv2.imshow("Orginal_img2",img2)

added = cv2.add(img1,img2)
cv2.imshow('Added Image',added)

added = cv2.addWeighted(img1,0.4,img2,0.5,0)
cv2.imshow('Weighted_add',added)

added = cv2.subtract(img1,img2,mask=None)
cv2.imshow('Sub_img',added)

added = cv2.bitwise_and(img1,img2,mask=None)
cv2.imshow('Bitwise_and',added)

added = cv2.bitwise_or(img1,img2,mask=None)
cv2.imshow('Bitwise_or',added)

added = cv2.bitwise_not(img1,img2,mask=None)
cv2.imshow('Bitwise_not',added)


cv2.waitKey(0)

    

    
```

```python
import cv2
import matplotlib.pyplot as plt
path = r'E:\Images\tajmahal.jpg'
img = cv2.imread(path,0) 
kernel1=np.array([[0,0,0],
                             [0,1,0],
                            [0,0,0]])
blur=cv2.filter2D(img,-1,kernel1)
laplacian=cv2.Laplacian(blur,5,cv2.CV_64F)
filtered_image=cv2.convertScaleAbs(laplacian)
plt.figure(figsize=(18,19))
plt.subplot(121)
plt.imshow(filtered_image,cmap='gray')
plt.title('laplacian')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = r'E:\Images\baby2.jpg'
img=cv2.imread(path)
kernel=np.ones((5,5),np.uint8)
img_erosion=cv2.erode(img,kernel,iterations=1)
img_dilation=cv2.dilate(img,kernel,iterations=1)
opening=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel,iterations=1)
closing=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel,iterations=1)
cv2.imshow("ORGINAL",img)   
cv2.imshow("EROSION",img_erosion)   
cv2.imshow("DILATION",img_dilation)   
cv2.imshow("OPENING",opening)   
cv2.imshow("CLOSING",closing)   
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
path = r'E:\Images\bike.jpg'
img=cv2.imread(path,0)
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
#canny
img_canny = cv2.Canny(img,100,200)
#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
cv2.imshow("Original Image", img)
cv2.imshow("Canny", img_canny)
cv2.imshow("Sobel X", img_sobelx)
cv2.imshow("Sobel Y", img_sobely)
cv2.imshow("Sobel", img_sobel)
cv2.imshow("Prewitt X", img_prewittx)
cv2.imshow("Prewitt Y", img_prewitty)
cv2.imshow("Prewitt", img_prewittx + img_prewitty)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
