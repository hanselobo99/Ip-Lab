```python
#10 a. Edge based Segmentation
import cv2
import numpy as np
path = r'/home/hansel/ip/Images/image7.jpg'
img=cv2.imread(path,0)
img = cv2.resize(img,(300,400))
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
#canny
img_canny = cv2.Canny(img,100,200)
#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
#roberts
kernela=np.array([[-1,0,1],[-1,0,1]],dtype=float)
kernelb=np.array([[[-1,0,1],[0,0,0]],dtype=float)
x=cv2.filter2D(img_gaussian,cv2.CV_16S,kernela)
y=cv2.filter2D(img_gaussian,cv2.CV_16S,kernelb)
absX=cv2.convertScaleAbs(x)
absY=cv2.convertScaleAbs(y)
Roberts=cv2.addWeighted(absX,0.5,absY,0.5,0)
cv2.imshow("Original Image", img)
cv2.imshow("Canny", img_canny)
cv2.imshow("Sobel X", img_sobelx)
cv2.imshow("Sobel Y", img_sobely)
cv2.imshow("Sobel", img_sobel)
cv2.imshow("Prewitt X", img_prewittx)
cv2.imshow("Prewitt Y", img_prewitty)
cv2.imshow("Prewitt", img_prewittx + img_prewitty)
cv2.imshow("Roberts",Roberts)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# a) Edge based Segmentation
#RegionBased
import cv2
import numpy as np
path = r'/home/hansel/ip/Images/image7.jpg'
img=cv2.imread(path,0)
img = cv2.resize(img,(300,400))
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
img_canny = cv2.Canny(img,100,200)
#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
cv2.imshow("Original Image", img)
cv2.imshow("Canny", img_canny)
cv2.imshow("Sobel X", img_sobelx)
cv2.imshow("Sobel Y", img_sobely)
cv2.imshow("Sobel", img_sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# b. Watershed based Segmentation
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image7.jpg'
img=cv.imread(path)
img = cv.resize(img,(300,400))
cv.imshow("Original image",img)

b,g,r = cv.split(img)
rgb_img = cv.merge([r,g,b])
cv.imshow("rgb image",rgb_img)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(gray,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
# # noise removal
kernel = np.ones((3,3),np.uint8)
opening = cv.morphologyEx(thresh,cv.MORPH_OPEN,kernel, iterations = 2)
cv.imshow("image without Noise",opening)
# sure background area
sure_bg = cv.dilate(opening,kernel,iterations=3)
cv.imshow("sure background area",sure_bg)
# Finding sure foreground area
dist_transform = cv.distanceTransform(opening,cv.DIST_L2,5)
ret, sure_fg = cv.threshold(dist_transform,0.7*dist_transform.max(),255,0)
cv.imshow("sure foreground area",sure_fg)
cv.waitKey(0)
cv2.destroyAllWindows()
```
