```python
# 3 Write a python program to demonstrate translation of the image from present position.

import cv2
import matplotlib.pyplot as plt
import numpy as np
path = r'/home/hansel/ip/Images/image3.jpg'
img=cv2.imread(path)
(rows,cols)=img.shape[:2]
A=np.float32([[1,0,100],[0,1,50]])
res=cv2.warpAffine(img,A,(rows,cols))
plt.subplot(121),plt.title("ORGINAL"),plt.imshow(img)
plt.subplot(122),plt.title("Translation"),plt.imshow(res)
```

```python
# 4 Write a python program to Enhance an image using Image
# Arithmetic and Logical Operations

import cv2
import matplotlib.pyplot as plt
image1 = cv2.imread(r"/home/hansel/ip/Images/image2.jpg")
image2 = cv2.imread(r"/home/hansel/ip/Images/image4.jpg")
img1 = cv2.resize(image1,(300,400))
img2 = cv2.resize(image2,(300,400))
cv2.imshow("Orginal_img1",img1)
cv2.imshow("Orginal_img2",img2)

add = cv2.add(img1,img2)
cv2.imshow('Added Image',add)

addWeighted = cv2.addWeighted(img1,0.4,img2,0.5,0)
cv2.imshow('Weighted_add',addWeighted)

subtract = cv2.subtract(img1,img2,mask=None)
cv2.imshow('Sub_img',subtract)

bitwise_and = cv2.bitwise_and(img1,img2,mask=None)
cv2.imshow('Bitwise_and',bitwise_and)

bitwise_or = cv2.bitwise_or(img1,img2,mask=None)
cv2.imshow('Bitwise_or',bitwise_or)

bitwise_not = cv2.bitwise_not(img1,img2,mask=None)
cv2.imshow('Bitwise_not',bitwise_not)


cv2.waitKey(0)
```

```python
# 7 Write a python program for gray level slicing with or without background
#withbackground
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread(r"/home/hansel/ip/Images/image2.jpg",0)
row, column= img.shape
img1 = np.zeros((row,column),dtype = 'uint8')
for i in range(row):
    for j in range(column):
        if(img[i,j]>50 and img[i,j]<150):
            img1[i,j] = 255
        else:
            img1[i,j] = img[i][j]
plt.title('Grey level slicing with background')
plt.imshow(img1,cmap="gray")
plt.show()
```

```python
# 7 Write a python program for gray level slicing with or without background
#withoutbackground
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread(r"/home/hansel/ip/Images/image2.jpg",0)
row, column= img.shape
img1 = np.zeros((row,column),dtype = 'uint8')
for i in range(row):
    for j in range(column):
        if(img[i,j]>10 and img[i,j]<60):
            img1[i,j] = 255
        else:
            img1[i,j] = 0
plt.title('Grey level slicing without background')
plt.imshow(img1,cmap="gray")
plt.show()
```

```python
# 8 Analyze a colour and gray scale image using histogram and interpret the same and display both histogram.
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread(r"/home/hansel/ip/Images/image7.jpg")
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist=plt.hist(imgGray.ravel(),256,[0,256])
hist1=cv2.calcHist([imgGray],[0],None,[256],[0,256])
plt.subplot(221),plt.imshow(imgGray,cmap="gray")
plt.subplot(222),plt.plot(hist1)
plt.xlim([0,256])
plt.show()


hist=plt.hist(img.ravel(),256,[0,256])
hist1=cv2.calcHist([img],[0],None,[256],[0,256])
hist2=cv2.calcHist([img],[1],None,[256],[0,256])
hist3=cv2.calcHist([img],[2],None,[256],[0,256])
plt.subplot(221),plt.imshow(img)
plt.subplot(222),plt.plot(hist1),plt.plot(hist2),plt.plot(hist3)
plt.xlim([0,256])
plt.show()
```

```python
# 9 Write a python program for enhance the brightness and contrast using histogram equalization.
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = r"/home/hansel/ip/Images/image7.jpg"
img=cv2.imread(path)
yuv=cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
yuv[:,:,0]=cv2.equalizeHist(yuv[:,:,0])
equalhist=cv2.cvtColor(yuv,cv2.COLOR_YUV2BGR)
images=[img,yuv,equalhist]
titles=["o","y","e"]
count=3
for i in range(0,count):
    plt.subplot(1,3,i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
```

```python
# 11 Demonstrate morphological operations in an input image and display the same.
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = r"/home/hansel/ip/Images/image7.jpg"
img=cv2.imread(path)
img = cv2.resize(img,(300,400))
kernel=np.ones((5,5),np.uint8)
img_erosion=cv2.erode(img,kernel,iterations=1)
img_dilation=cv2.dilate(img,kernel,iterations=1)
opening=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel,iterations=1)
closing=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel,iterations=1)
cv2.imshow("ORGINAL",img)   
cv2.imshow("EROSION",img_erosion)   
cv2.imshow("DILATION",img_dilation)   
cv2.imshow("OPENING",opening)   
cv2.imshow("CLOSING",closing)   
cv2.waitKey(0)
```

```python
# 13 Demonstrate Adaptive thresholding in an input image. Use both the Adaptive thresholding methods.
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = r"/home/hansel/ip/Images/image7.jpg"
img=cv2.imread(path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
thresh1=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,199,5)
thresh2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,199,5)
plt.subplot(121),plt.title("Adm"),plt.imshow(thresh1,cmap="gray")
plt.subplot(122),plt.title("Adg"),plt.imshow(thresh2,cmap="gray")
```

```python
# 14 Apply Otsu thresholding in an input image with binary thresholding and Inverted binary threshold.
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = r"/home/hansel/ip/Images/image7.jpg"
img=cv2.imread(path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh1=cv2.threshold(img,120,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
ret,thresh2=cv2.threshold(img,120,255,cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
plt.subplot(121),plt.title("Otsub"),plt.imshow(thresh1,cmap="gray")
plt.subplot(122),plt.title("Otsui"),plt.imshow(thresh2,cmap="gray")
```
