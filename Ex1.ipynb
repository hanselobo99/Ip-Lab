```python
# e. Display the image such that the pixels having intensity values below than 50 will  display as black and pixels having intensity values above than 150 will display as white. And the pixels between these will display as it is.

import cv2
import numpy as np
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image1.jpg'
image1 = cv2.imread(path)
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image,(300,400))
image1 = cv2.resize(image1,(300,400))
h = image.shape[0]
w = image.shape[1]
plt.imshow(image1),plt.show()

for i in range (0,h):
    for j in range (0,w):
        pix = image[i,j]
        if pix > 150:
            image1[i,j] = np.array([255,255,255])
        elif pix < 50:
            image1[i,j] = np.array([0,0,0])
plt.imshow(image1)


```

```python
# b. Increase and decrease the brightness and
# Save it in a drive

import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
directory='/home/hansel/ip/Images/drive1/'
try:
    os.mkdir(directory)
except Exception as e:
    print("Save in Drive")
path = r'/home/hansel/ip/Images/image7.jpg'
image = cv2.imread(path)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
value=30
vValue = hsv[...,2]
hsv[...,2] = np.where((255-vValue)<value,255,vValue+value)
plt.subplot(131),plt.title("Original"),plt.imshow(image), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)),
plt.title('brightened image'), plt.xticks([]), plt.yticks([])
cv2.imwrite(directory+"file1.jpg",cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB))
value=-30
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
vValue = hsv[...,2]
hsv[...,2] = np.where((255-vValue)<value,255,vValue+value)
plt.subplot(133),plt.imshow(cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)),
plt.title('Decreased brightness'), plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite(directory+"file2.jpg",cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB))
cv2.waitKey(0)
```
