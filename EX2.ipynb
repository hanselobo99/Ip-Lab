```python
# a. Read a colour image, convert it into gray scale
# image. Increase and decrease the contrast and
# display all the three images and save.
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
directory='/home/hansel/ip/Images/drive2/'
try:
    os.mkdir(directory)
except Exception as e:
    print("Save in Drive")
path = r'/home/hansel/ip/Images/image3.jpg'
image = cv2.imread(path,0)
plt.subplot(131),plt.title("Actual Image"),plt.imshow(image,cmap="gray"),plt.xticks([]), plt.yticks([])

intensity_matrix=np.ones(image.shape,dtype="uint8")*60
bi=cv2.add(image,intensity_matrix)
di=cv2.subtract(image,intensity_matrix)
plt.subplot(132),plt.title("Bright"),plt.imshow(bi,cmap="gray"),plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.title("Dark"),plt.imshow(di,cmap="gray"),plt.xticks([]), plt.yticks([])
cv2.imwrite(directory+"file1.jpg",bi)
cv2.imwrite(directory+"file2.jpg",di)
cv2.waitKey(0)
```

```python
# b. Downscale the image by resizing with custom
# height and width while not preserving the
# aspect ratio.
import cv2
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image3.jpg'
img=cv2.imread(path)
down_width=300
down_height=200
down_points=(down_width,down_height)
dscale=cv2.resize(img,down_points,interpolation=cv2.INTER_LINEAR)
plt.subplot(121),plt.title("Orginal"),plt.imshow(img)
plt.subplot(122),plt.title("DOWN SCALE"),plt.imshow(dscale)
```

```python
# c. Scale the image vertically and horizontally and
# display all three images.
import cv2
import matplotlib.pyplot as plt
path = r'/home/hansel/ip/Images/image3.jpg'
img=cv2.imread(path)
new_width=1300
new_height=2100
hz=(new_width,img.shape[0])
vr=(new_height,img.shape[1])
dh=cv2.resize(img,hz,interpolation=cv2.INTER_AREA)
dv=cv2.resize(img,vr,interpolation=cv2.INTER_AREA)
plt.subplot(131),plt.title("Orginal"),plt.imshow(img)
plt.subplot(132),plt.title("Horizontal"),plt.imshow(dh)
plt.subplot(133),plt.title("Vertical"),plt.imshow(dv)
```

```python
# d. Downsampling and Upsampling an image
# while preserving the aspect ratio.
import cv2
import matplotlib.pyplot as plt
path = r'/home/hansel/ip/Images/image3.jpg'
img=cv2.imread(path)
ds=cv2.resize(img,(0,0),fx=0.5,fy=0.5)
us=cv2.resize(img,(0,0),fx=2,fy=2)
plt.subplot(131),plt.title("Orginal"),plt.imshow(img)
plt.subplot(132),plt.title("Downsampling"),plt.imshow(ds)
plt.subplot(133),plt.title("Upsampling"),plt.imshow(us)
```

```python
# e. Resize the image by stretching, shrinking and
# zooming the same by using interpolation
# methods.
import cv2
import matplotlib.pyplot as plt
path = r'/home/hansel/ip/Images/image3.jpg'
img=cv2.imread(path)
i_zoom=cv2.resize(img,(720,600),interpolation=cv2.INTER_AREA)
i_strech=cv2.resize(img,(600,600),interpolation=cv2.INTER_NEAREST)
i_shrink=cv2.resize(img,(350,350),interpolation=cv2.INTER_LINEAR)
plt.title("Original"),plt.imshow(img)
plt.show()
plt.subplot(131),plt.title("Zoom"),plt.imshow(i_zoom)
plt.subplot(132),plt.title("Strech"),plt.imshow(i_strech)
plt.subplot(133),plt.title("Shrink"),plt.imshow(i_shrink)
```

```python
# f. Rotate the image to about 60 degree without scaling.
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = r'/home/hansel/ip/Images/image3.jpg'
img=cv2.imread(path)
img = cv2.resize(img,(400,400))
[rows,cols]=img.shape[:2]
M=cv2.getRotationMatrix2D((cols/2,rows/2),60,1)
res=cv2.warpAffine(img,M,[rows,cols])
plt.subplot(121),plt.title("ORGINAL"),plt.imshow(img)
plt.subplot(122),plt.title("Rotation"),plt.imshow(res)
```

```python
# g. Get a better information about an image by changing the perspective of an image
import cv2
import numpy as np
import matplotlib.pyplot as plt
width,height=200,300
path = r'/home/hansel/ip/Images/image8.jpeg'
image=cv2.imread(path)
pts1=np.float32([[30,450],[600,40],[350,900],[960,480]])
pts2=np.float32([[0,0],[width,0],[0,height],[width,height]])
matrix=cv2.getPerspectiveTransform(pts1,pts2)
output=cv2.warpPerspective(image,matrix,(width,height))  
plt.title("original")
plt.imshow(image)
plt.show()
plt.title("output")
plt.imshow(output)
plt.show()
```
