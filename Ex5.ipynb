```python
# a. Read a gray scale image
import cv2
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image6.jpg'
img=cv2.imread(path) 
img2=cv2.imread(path,0) 
plt.title("Original"),plt.imshow(img),plt.show()
plt.title("Gray SCale"),plt.imshow(img2,cmap="gray")    
```

```python
# b. Apply Custom 2D-Convolution 3 X 3 Kernel,combine two images and display.
import cv2
import numpy as np
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image6.jpg'
img=cv2.imread(path)
kernel1=np.array([[0,0,0],
                 [0,1,0],
                 [0,0,0]])
identity=cv2.filter2D(src=img,ddepth=-1,kernel=kernel1)
plt.title("Combined"),plt.imshow(np.hstack([img,identity])),plt.show()

```

```python
# c. Apply Mean/Box/Average Filter, concatenate
# two images horizontally and display.
import cv2
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image6.jpg'
img = cv2.imread(path,0)
meanf = cv2.boxFilter(img,-1,(2,2))
addi=cv2.hconcat([img,meanf])
plt.title("concatenated"),plt.imshow(addi,cmap="gray"),plt.show()
```

```python
# d. Apply Gaussian blurring, concatenate two
# images vertically and display.
import cv2
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image6.jpg'
img = cv2.imread(path)
img = cv2.resize(img,(300,300))
gblur = cv2.GaussianBlur(img,(7,7),0)
vis = cv2.vconcat([img,gblur])
plt.title("Gaussian blur"),plt.imshow(vis),plt.show()
```

```python
# e. Apply Median blurring and display the image
import cv2
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image6.jpg'
img = cv2.imread(path)
mblur = cv2.medianBlur(img,5)
plt.subplot(121),plt.title("Original"),plt.imshow(img)
plt.subplot(122),plt.title("Median Blur"),plt.imshow(mblur) 
```

```python
# f. Apply Bilateral blurring and display the image
import cv2          
from matplotlib import pyplot as plt
path = r'/home/hansel/ip/Images/image6.jpg'
img = cv2.imread(path)      
bblur = cv2.bilateralFilter(img,9,75,75) 
plt.subplot(121),plt.title("Original"),plt.imshow(img)
plt.subplot(122),plt.title("Bilateral blur"),plt.imshow(bblur)   
```
